```python
import numpy as np
import matplotlib.pyplot as plt

def plot_img(id):
  arr = []
  with open(f"img/{id}", "r") as f:
    for line in f:
      arr.append(float(line))

  arr = np.array(arr).reshape((512, 512))
  plt.xticks([])
  plt.yticks([])
  plt.imshow(arr, "GnBu")

id = 0
```

```python
plot_img(id)
id += 1
```
