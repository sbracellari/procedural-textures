```python
import numpy as np
import matplotlib.pyplot as plt

arr = []
with open("data.txt", "r") as f:
  for line in f:
    arr.append(float(line))

arr = np.array(arr).reshape((4000, 4000))
plt.imshow(arr, "GnBu")
```
